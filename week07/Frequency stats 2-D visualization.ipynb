data set is in [Homework_1_Data.txt](Homework_1_Data.txt). Each data point is a two-dimensional vector, $\mathbf{x} = (x_1, x_2)$.

### Part (A): Visualization and Interpretation

- Make a 2-D visualization the distribution of the data. 

- Visualize the the pdf, $f_X$, of the underlying distribution of the data.

- Visualize the distribution defined by $f_{x_2 \mid x_1}$ for $x_1 \in [3.99, 4.01]$.

- Visualize the distribution defined by $f_{x_1}$.


### Part (B): Estimation

- Empirically estimate the mean of the distribution $f_{x_1}$. Estimate, also the SE (standard error) of the estimate.

- Empirically estimate the standard deviation of the distribution $f_{x_2 \mid x_1}$, for $x_1 \in [3.99, 4.01]$. Estimate, also the SE (standard error) of the estimate.

- Given the SE, How many digits in your standard deviation estimate are significant? Explain why.

```python
# load data into 2-d array
import numpy as np
import matplotlib.pyplot as plt
arr = np.loadtxt('Homework_1_Data.txt', dtype = float, delimiter = ',')
```

## Make a 2-D visualization the distribution of the data. 

```python
plt.plot(arr[:,0],arr[:,1],'.')
plt.figure(figsize=(20,10))
plt.style.use('ggplot')
plt.show()
```

## Visualize the the pdf, $f_X$, of the underlying distribution of the data.


```python
# plot histgram
from matplotlib.colors import LogNorm
plt.hist2d(arr[:,0],arr[:,1], bins=20
           #norm=LogNorm(), 
          )

plt.colorbar()
plt.show()
```

##  Visualize the distribution defined by $f_{x_2 \mid x_1}$ for $x_1 \in [3.99, 4.01]$.


```python
# take indices where x1 is in the range:
index = np.logical_and(arr[:,0] > 3.99, arr[:,0] < 4.01)

# take all points where x1 is in the range
f_x2 = arr[index,:]

plt.hist(f_x2[:,0])
plt.style.use('ggplot')
plt.show()


# use seaborn
import seaborn as sns
#sns.set_style("whitegrid")
#sns.set_context("poster")
sns.distplot(f_x2[:,0])
plt.show()
```

## - Visualize the distribution defined by $f_{x_1}$.


```python
x1 = arr[:,0]
sns.distplot(x1)
plt.show()
```

## - Empirically estimate the mean of the distribution $f_{x_1}$. Estimate, also the SE (standard error) of the estimate.

## - Empirically estimate the standard deviation of the distribution $f_{x_2 \mid x_1}$, for $x_1 \in [3.99, 4.01]$. Estimate, also the SE (standard error) of the estimate.

## - Given the SE, How many digits in your standard deviation estimate are significant? Explain why.

```python
mu1 = x1.mean()
std1 = x1.std()

x2c = f_x2[:,1]
mu2c = x2c.mean()
std2c= x2c.std()
print(mu1, std1)
print(mu2c, std2c)
```
